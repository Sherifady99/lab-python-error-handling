````python
# Creating Product List

products = ["t-shirt", "mug", "hat", "book", "keychain"]
````

````python
# Applying error handling to initialize inventory function

def initialize_inventory(products):
    inventory = {}
    for product in products:
        valid_quantity = False
        while not valid_quantity:
            try:
                quantity = int(input(f"Enter the quantity of {product}s available: "))
                if quantity < 0:
                    raise ValueError("Invalid quantity! Please enter a non-negative value.")
                valid_quantity = True
            except ValueError as error:
                print(f"Error: {error}")
        inventory[product] = quantity
    return inventory

inventory = initialize_inventory(products)

print(inventory)
````

````python
def get_customer_orders(products, inventory):
    valid_num_orders = False
    while not valid_num_orders:
        try:
            number_of_orders = int(input("Enter the number of of customer orders: "))
            if number_of_orders <0:
                raise ValueError ("Number of orders can't be negative")
            if number_of_orders == 0:
                raise ValueError("Number of orders cannot be 0")
            valid_num_orders = True
        except ValueError as error:
            print(f"Error: {error}")

    customer_orders = set()
    for _ in range(number_of_orders):
        valid_product_name = False
        while not valid_product_name:
            try:
                product_name = input("Enter the name of a product (t-shirt, mug, hat, book, keychain): ")
                if product_name not in products:
                    raise ValueError ("This product does not exist in our inventory")
                elif inventory[product_name] <=0:
                    raise ValueError("We do not have any remaining stock for this product")
                else:
                    valid_product_name = True
                    customer_orders.add(product_name)
                    inventory[product_name] -= 1
            except ValueError as error:
                print(f"Error: {error}")
    return customer_orders

customer_orders = get_customer_orders(products, inventory)

print("Your final order is:", customer_orders)
````

````python
# Calculating total price with error handling

def calculate_total_price(customer_orders):
    total_price = 0
    for product in customer_orders:
        valid_price = False
        while not valid_price:
            try:
                price_input = float(input(f"Please enter the price of {product}: "))
                if price_input <0:
                    raise ValueError("Price cannot be negative")
                elif price_input == 0:
                    raise ValueError ("Are you sure you want to give this product away for free? :)")
                else:
                    total_price += price_input
                    valid_price = True
            except ValueError as error:
                print(f"Error: {error}")
    return total_price

total_price = calculate_total_price(customer_orders)
print("Total price for your order is:", total_price)
````
